**- 1st request: Get each column mean `np.mean(column)`, median `np.median(column)`, range value `np.amax(column) - np.amin(column)`, standard deviation `np.std(column)`**

```python
col_names = ["Home Value", "HH Inc", "Per Cap Inc", "Pct Owner Occ"]
mean_values = []
median_values = []
std_values = []
range_values = []

for i in range(1, 5):
    temp = df.iloc[:, i]
    tmp = float(format(np.mean(temp), '.2f'))
    mean_values.append(tmp)
    median_values.append(np.median(temp))
    tmp = float(format(np.std(temp, ddof = 1), '.2f'))
    std_values.append(tmp)
    tmp = float(format(np.amax(temp) - np.amin(temp), '.2f'))
    range_values.append(tmp)

calculation = [mean_values, median_values, range_values, std_values]
df2 = pd.DataFrame(data = calculation, columns = col_names)
df2["Calculation"] = ["mean", "median", "range", "std"]
cols = ["Calculation", "Home Value", "HH Inc", "Per Cap Inc", "Pct Owner Occ"]
df2 = df2[cols]
df2
```

**- 2nd & 3rd requests: draw Frequency Histogram and Boxplot for each variable**

+ use `np.histogram(column)` to conveniently take data

+ use `plt.hist()` to draw Frenquency Histograms

+ use `plt.boxplot()` to draw Boxplots

+ use other functions to customize the charts

```python
x = df["Home Value"]
counts, bins = np.histogram(df["Home Value"])
ytick = [0, 4, 8, 12, 16, 20]
plt.figure(figsize = (12, 6))

plt.hist(x, bins, rwidth = 1, color ="#FF5B5B", edgecolor = "#000")
plt.xticks(bins)
plt.yticks(ytick)
plt.title("Home Value Frequency Histogram")
plt.xlabel("Home Value ($)")
plt.ylabel("Frequency (time)")
plt.grid(axis = "y", linewidth = 0.5)
for i in range(len(counts)):
    height = int(counts[i])
    plt.annotate(height,
                 xy = (bins[i], height),
                 textcoords = "offset points",
                 xytext = (31, 3),
                 ha = 'center', va = 'bottom'
                )
plt.show()

plt.figure(figsize = (12, 6))
plt.boxplot(x)
plt.xticks([])
plt.title("Home Value Boxplot")
plt.xlabel("Home Value")
plt.ylabel("Currency ($)")
plt.grid(axis = "y", linewidth = 0.5)
plt.show()
```

**- 4th request: Draw scatter plots (2 variables/ 1 chart) to compare and find relationship among the variables**

+ use subplot (2x3) to store all 6 scatter plots (more convenient while compare and find insights)

+ use `subplot(pos).scatter()` to fill each subplot with a scatter chart

+ `set_xlabel()`, `set_ylabel()`: set x & y information

+ `set_title()`: set the chart's header

+ some other parameters and functions to customize the plots (`fontsize`, `tick_params()`, ...)

```python
fig, ((ax1, ax2,ax3),(ax4, ax5,ax6)) = plt.subplots(nrows=2, ncols=3,figsize=(30,16))

a = df["Home Value"]
b = df["HH Inc"]
c = df["Per Cap Inc"]
d = df["Pct Owner Occ"]

ax1.scatter(b, a, color = "#FF5B5B")
ax1.set_title("Home Value & HH Inc scatterplot", fontsize=18)
ax1.set_xlabel("HH Inc ($)", fontsize=16)
ax1.set_ylabel("Home Value ($)", fontsize=16)
ax1.tick_params(axis='both', which='major', labelsize=14)


ax2.scatter(c, a, color = "#1ABC9C")
ax2.set_title("Home Value & Per Cap Inc Inc scatterplot", fontsize=18)
ax2.set_xlabel("Per Cap Inc ($)", fontsize=16)
ax2.set_ylabel("Home Value ($)", fontsize=16)
ax2.tick_params(axis='both', which='major', labelsize=14)


ax3.scatter(a, d, color = "#FF5B5B")
ax3.set_title("Home Value & Pct Owner Occ scatterplot", fontsize=18)
ax3.set_ylabel("Pct Owner Occ (percent)", fontsize=16)
ax3.set_xlabel("Home Value ($)", fontsize=16)
ax3.tick_params(axis='both', which='major', labelsize=14)

ax4.scatter(c, b, color = "#5DADE2")
ax4.set_title("HH Inc & Per Cap Inc scatterplot", fontsize=18)
ax4.set_xlabel("Per Cap Inc ($)", fontsize=16)
ax4.set_ylabel("HH Inc ($)", fontsize=16)
ax4.tick_params(axis='both', which='major', labelsize=14)

ax5.scatter(b, d, color = "#F4D03F")
ax5.set_title("HH Inc & Pct Owner Occ scatterplot", fontsize=18)
ax5.set_ylabel("Pct Owner Occ (percent)", fontsize=16)
ax5.set_xlabel("HH Inc ($)", fontsize=16)
ax5.tick_params(axis='both', which='major', labelsize=14)

ax6.scatter(c, d, color = "#5DADE2")
ax6.set_title("Per Cap Inc & Pct Owner Occ scatterplot", fontsize=18)
ax6.set_ylabel("Pct Owner Occ (percent)", fontsize=16)
ax6.set_xlabel("Per Cap Inc ($)", fontsize=16)
ax6.tick_params(axis='both', which='major', labelsize=14)


plt.show()
```

___

**- 5th request: Calculate correlation coefficient and find the relationships among the variables**

```python
ab = float(format(np.corrcoef(a,b)[0][1], '.4f'))
ac = float(format(np.corrcoef(a,c)[0][1], '.4f'))
ad = float(format(np.corrcoef(a,d)[0][1], '.4f'))
bc = float(format(np.corrcoef(b,c)[0][1], '.4f'))
bd = float(format(np.corrcoef(b,d)[0][1], '.4f'))
cd = float(format(np.corrcoef(c,d)[0][1], '.4f'))
row1 = [1, ab, ac, ad]
row2 = [ab, 1, bc, bd]
row3 = [ac, bc, 1, cd]
row4 = [ad, bd, cd, 1]
rows = [row1, row2, row3, row4]
colNames = ["Home Value", "HH Inc", "Per Cap Inc", "Pct Owner Occ"]
df3 = pd.DataFrame(data = rows, columns = colNames)
df3["Value"] = ["Home Value", "HH Inc", "Per Cap Inc", "Pct Owner Occ"]
colNames = ["Value","Home Value", "HH Inc", "Per Cap Inc", "Pct Owner Occ"]
df3 = df3[colNames]
df3
```
